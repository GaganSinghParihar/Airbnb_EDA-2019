```python
# --------------------------------------------
# Step 2: Data Cleaning
# --------------------------------------------

# Import required libraries
import pandas as pd
import numpy as np

# Load the dataset
# Using a relative path ensures compatibility across systems.
df = pd.read_csv("../data/airbnb_nyc_2019.csv")

# -------------------------
# 1. Handle Missing Values
# -------------------------

# Convert last_review column to datetime format.
# Some values may be missing, so errors='coerce' handles invalid formats safely.
df['last_review'] = pd.to_datetime(df['last_review'], errors='coerce')
```

```python
# Missing values in reviews_per_month occur when a listing has no reviews.
# We logically replace missing values with 0.
df['reviews_per_month'] = df['reviews_per_month'].fillna(0)

# name and host_name have very few missing values.
# Since names do not impact numerical analysis, we leave them as NaN for now.
```
